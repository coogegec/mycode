```python
import numpy as np
a=np.arange(0,10,0.5)
print(len(a),a) # 0 ~ 10까지 0.5 간격으로 값 생성
data=np.sin(a)
print(data) 
print(len(data))
```

```python
timestep=3 # t-3, t-2, t-1 데이터를 이용하여 t 시점의 값을 예측
X,y=[],[]

for i in range(timestep, len(data)):
    X.append(np.array([data[i-j] for j in range(timestep,0,-1)]))
    y.append(data[i])

print(X)
print(y)
```

```python
# 신경망에 입력하기 위해 차원 변경
X, y= np.array(X).reshape(len(X),3,1), np.array(y).reshape(len(y),1)
print(X.shape)
print(y.shape)
```

```python
from tensorflow.keras.models import Sequential
from keras.layers import SimpleRNN, Dense

np.random.seed(0)

model=Sequential()
model.add(SimpleRNN(units=10, input_shape=(3,1))) # units : 출력 노드수
model.add(Dense(5, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='sgd')
```

```python
history=model.fit(X, y, epochs=1000, verbose=1)
```

```python
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(15,4))
plt.plot(data[3:], 'ro-', label='real values') # 실제값
plt.plot(model.predict(X), 'bo-', label='predict values') # 예측값
plt.legend()
plt.show()
```
