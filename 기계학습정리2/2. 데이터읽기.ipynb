```python
# 엑셀 파일 읽기
# pip install xlrd
import pandas as pd
# 데이터 URL
url = 'https://tinyurl.com/simulated-excel'
# 데이터 적재
# sheet_name 시트의 인덱스, header=1 1번행을 헤더로 사용
df = pd.read_excel(url, sheet_name=0, header=1)
# 처음 두 행을 확인
df.head(2)
```

```python
# JSON 파일 적재하기
import pandas as pd
# 데이터 URL
url = 'https://tinyurl.com/simulated-json'
# 데이터 적재
df = pd.read_json(url)
# 처음 두 행을 확인
df.head(2)
```

```python
# SQL 데이터베이스로부터 적재
# pip install sqlalchemy
import pandas as pd
from sqlalchemy import create_engine
# 데이터베이스에 연결
conn = create_engine('sqlite:///d:/data/db/ontime.db')
# 데이터를 적재
df = pd.read_sql_query('SELECT * FROM ontime LIMIT 10', conn)
# 처음 두 개의 행을 확인
df.head(2)
```
