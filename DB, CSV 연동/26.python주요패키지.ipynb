```python
student1=[60,70,80]
student2=[77,88,99]
points=np.array([student1,student2])
print(points)
print(points.shape)
```

```python
print(points.sum()) # 전체 합계
print(points.sum(axis=0)) # 열의 합계
print(points.sum(axis=1)) # 행의 합계
print(points.mean()) # 전체 평균
print(points.mean(axis=0)) # 열의 평균
print(points.mean(axis=1)) # 행의 평균
```

```python
arr=np.maximum(student1,student2) # maximum : 각 요소 중 큰 값을 뽑아서 새로운 배열을 생성
print(arr)
print(arr.sum()) # 합계
print(arr.mean()) # 평균
print(np.sort(arr)) # 오름차순
print(np.sort(arr)[::-1]) # 내림차순 [::-1] - [start:stop:step]  모든 범위에서 step=-1
```

```python
a=pd.DataFrame({
    'name':['김철수','김대희','송영민','이희연','정미숙'],
    'weight':[80.0,70.4,65.5,45.9,51.2],
    'height':[170,180,155,143,154],
    'gender':['m','m','m','f','f']
})
print('체중')
print(a['weight'])
print('신장과 체중')
print(a[['weight','height']]) # 2개 이상의 시리즈를 출력할 때는 [[ ]]
```

```python
print('신장 170 이상')
print(a[a.height>=170])
print('남성')
print(a[a.gender=='m'])
print('키 오름차순')
print(a.sort_values(by='height'))
print('몸무게 내림차순')
print(a.sort_values(by='weight', ascending=False))
```

```python
df=pd.read_csv('d:/data/traffic/data.csv', encoding='ms949')
df.head()
```

```python
print(df2["1월"].mean()) # 열의 평균
print()
print(df2.sum(axis=0)) # 각 열의 합
print()
print(df2.head().sum(axis=1)) # 각 행의 합
print()
print(df2["1월"].var()) # 열의 분산
print()
f1=lambda x:x.max()-x.min() # x가 입력되면 최댓값에서 최솟값을 뺀 값을 리턴하는 람다식
print(df2.apply(f1,axis=0)) # 각 열에 f1 함수 적용
print()
print(df2.head().apply(f1,axis=1)) # 각 행에 f1 함수 적용
print()
```
