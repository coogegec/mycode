```python
# x or 함수
def xor(a,b):
    if a+b == 1:
        return 1
    else:
        return 0

print(xor(0,0)) # False
print(xor(0,1)) # True
print(xor(1,0)) # True
print(xor(1,1)) # False
```

```python
# 리스트의 마지막 원소를 맨 앞으로 이동시키는 함수
def rotate(items):
    result=[] # 빈 리스트에
    result.append(items[-1]) # 마지막 원소를 먼저 넣고
    result=result+items[:-1] # 마지막 원소를 제외한 모든 값들을 넣는다
    return result

a=[1,2,3,4]
rotate(a)
```

```python
def my_zip(a,b):
    if len(a)!=len(b):
        print('두 입력값의 길이가 다릅니다.')
    else:
        n=len(a)
        c=[]
        for i in range(n):
            c.append((a[i],b[i]))
        return c

a=[1,2,3]
b=['a','b','c']
my_zip(a,b)
```

```python
# 복리 계산
def saving(money, years, rate):
    result=money*(1+rate*0.01)**years
    return result
```

```python
# 만원을 연이율 3%, 5%, 7% 이율로 예금하였을 때 10년이 될 때까지 매년 총액수를 함수로
money=10000
years=np.arange(0,11)
rate=[3,5,7]
total_money=np.zeros((len(rate),len(years)))

for i in range(len(rate)):
    for j in range(len(years)):
        total_money[i,j]=saving(money,years[j],rate[i])
    plt.plot(years, total_money[i,:], label=f'{rate[i]}%')

plt.legend(fontsize=14)
plt.grid()
plt.xlabel('Years',fontsize=14)
plt.ylabel('Total money',fontsize=14)
```

```python
# 분산과 표준편차
x=[0,1,3,6,12,13,10,7,5,1]

mean=np.average(x)
print('평균 :',mean)
print('분산 :',np.var(x).1f)
print('표준편차 :',np.std(x))
```

```python
variance=0
for i in range(len(x)):
    variance+=(x[i]-mean)**2
variance/=len(x)
std=np.sqrt(variance)
print(variance)
print(std)
```

```python
# 주사위를 두 번 던져서 처음 나온 숫자를 x, 두번째 나온 숫자를 y라 할 때 x-y의 분산은?
x=np.arange(1,7)
y=np.arange(1,7)
x_minus_y=[]
for i in x:
    for j in y:
        x_minus_y.append(i-j)

print(x_minus_y)
print(np.mean(x_minus_y))
print(f'{np.var(x_minus_y):.3f}')
```

```python
# 아래와 같은 2개의 히스토그램을 보고 평균과 표준편차가 큰 쪽을 고르시오.
bin=np.arange(-120,120,10)
hist1=np.random.normal(0,30,1000) # normal(평균,표준편차,샘플개수) : 정규분포 난수 생성
hist2=np.random.normal(10,20,1000)
plt.hist(hist1,bin,alpha=0.5,rwidth=0.8,color="blue")
plt.hist(hist2,bin,alpha=0.5,rwidth=0.8,color="yellow")

print(f'{np.mean(hist1):.3f}')
print(f'{np.mean(hist2):.3f}')
print(f'{np.std(hist1):.3f}')
print(f'{np.std(hist2):.3f}')
```
