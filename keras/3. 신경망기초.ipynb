```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
#  
x=np.array([0,1,2,3,4]) #    
y=2*x+1 #     *x + 
print("x :",x)
print("y :",y)
#  
model=Sequential() #   
model.add(Dense(1,input_shape=(1,))) #   1     
model.compile(optimizer='sgd', loss='mse') #  
# SGD :   (Stochastic Gradient Descent), mse : ,   

model.fit(x[:2].reshape(-1,1), y[:2].reshape(-1,1), epochs=500, verbose=1) # model.fit(, )
# x[:2] : 0, 1 2 
```

```python
#  
mse=model.evaluate(x[2:], y[2:])
print(mse)
print(' :', y[2:])
print(' :', model.predict(x[2:]).flatten())
```
