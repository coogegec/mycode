```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
# 입출력 데이터
x=np.array([0,1,2,3,4]) # 딥러닝 모델에 입력할 값
y=2*x+1 # 출력값    가중치*x + 편향
print("x :",x)
print("y :",y)
# 인공신경망 모델링
model=Sequential() # 케라스 인공신경망 생성
model.add(Dense(1,input_shape=(1,))) # 입력 노드 1개를 가지는 선형 처리 계층 추가
model.compile(optimizer='sgd', loss='mse') # 모델 컴파일
# SGD : 확률적 경사 하강법(Stochastic Gradient Descent), mse : 손실함수, 평균 제곱 오차

model.fit(x[:2].reshape(-1,1), y[:2].reshape(-1,1), epochs=500, verbose=1) # model.fit(입력데이터, 출력데이터)
# x[:2] : 0, 1번 2개의 데이터
```

```python
# 성능 평가
mse=model.evaluate(x[2:], y[2:])
print(mse)
print('실제값 :', y[2:])
print('예측값 :', model.predict(x[2:]).flatten())
```
