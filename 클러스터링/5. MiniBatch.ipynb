```python
from sklearn.datasets import make_blobs

X, y=make_blobs(n_samples=150000, cluster_std=[1.0, 2.5, 0.5], random_state=170)
# n_samples : 샘플 개수
```

```python
# 기본 모형
from sklearn.cluster import KMeans

model1=KMeans(n_clusters=3).fit(X)
```

```python
# 미니배치 적용한 모형 - 속도가 더 빠름
from sklearn.cluster import MiniBatchKMeans

model2=MiniBatchKMeans(n_clusters=3, batch_size=1500, compute_labels=True).fit(X)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

idx=np.random.randint(150000, size=300)
plt.subplot(121)
plt.scatter(X[idx,0], X[idx,1], c=model1.labels_[idx])
plt.title('KMeans')
plt.subplot(122)
plt.scatter(X[idx,0], X[idx,1], c=model2.labels_[idx])
plt.title('MiniBatch KMeans')
plt.tight_layout()
plt.show()
# 결과에는 큰 차이가 없다
```
