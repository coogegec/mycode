```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html=urlopen('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25ED%2595%25A9%25EB%25A6%25AC%25EC%25A0%2581&searchPcondition=1&searchCategory=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&collName=KORBOOK&from_CollName=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&searchOrder=0&vPstrTab=PRODUCT&from_coll=KORBOOK&currentPage=1&orderClick=LAH')
bs=BeautifulSoup(html,'html.parser')

titles=bs.select('div.title > a > strong')
authors=bs.select('div.author > a:nth-child(1)')
prices=bs.select('div.org_price')
books=list(zip(titles, authors, prices))
print(len(books),"권이 검색되었습니다.")
print('===============================================')
for book in books:
    print(book[0].text.strip(),'-',book[1].text.strip(),'-',book[2].text.strip())
```

```python
items=[]
for book in books:
    items.append([book[0].text.strip(),book[1].text.strip(),book[2].text.strip()])
```

```python
import pandas as pd
df=pd.DataFrame(items, columns=['도서명','저자','정가'])
df.head()
```

```python
df.to_csv('d:/data/book/data.csv', encoding='ms949', index=False)
```

```python
import math

html=urlopen('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25ED%2595%25A9%25EB%25A6%25AC%25EC%25A0%2581&searchPcondition=1&searchCategory=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&collName=KORBOOK&from_CollName=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&searchOrder=0&vPstrTab=PRODUCT&from_coll=KORBOOK&currentPage=1&orderClick=LAH')
bs=BeautifulSoup(html,'html.parser')

total=bs.select_one('#searchCategory_1 > small')
pages=int(total.text.strip().replace(',',''))/20
pages=math.ceil(pages)
print(pages)
```

```python
import time

total_titles=[]
total_authors=[]
total_prices=[]
for i in range(1, pages+1):
    html=urlopen(f'https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25ED%2595%25A9%25EB%25A6%25AC%25EC%25A0%2581&searchPcondition=1&searchCategory=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&collName=KORBOOK&from_CollName=%EA%B5%AD%EB%82%B4%EB%8F%84%EC%84%9C@KORBOOK&searchOrder=0&vPstrTab=PRODUCT&from_coll=KORBOOK&currentPage={i}&orderClick=LAH')
    bs=BeautifulSoup(html,'html.parser')
    titles=bs.select('div.title > a > strong')
    authors=bs.select('div.author > a:nth-child(1)')
    prices=bs.select('div.org_price')
    total_titles+=titles
    total_authors+=authors
    total_prices+=prices
    time.sleep(0.1)
books=list(zip(total_titles, total_authors, total_prices))
print(len(books),"권이 검색되었습니다.")
print('===============================================')
for book in books:
    print(book[0].text.strip(),'-',book[1].text.strip(),'-',book[2].text.strip())
```
