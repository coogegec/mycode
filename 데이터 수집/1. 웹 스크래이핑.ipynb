```python
from bs4 import BeautifulSoup
import urllib.request as req
url="http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
res=req.urlopen(url)
bs=BeautifulSoup(res, 'html.parser')
title=bs.find('title').text
wf=bs.find('wf').text
print(title)
print(wf)

result1=[]
result2=[]
result3=[]
result4=[]
result5=[]
items=bs.find_all('data')
for item in items:
    times=item.find_all('tmef')
    wfs=item.find_all('wf')
    tmns=item.find_all('tmn')
    tmxs=item.find_all('tmx')
    rnsts=item.find_all('rnst')
    for idx in range(len(times)):
        result1.append(times[idx].text)
        result2.append(wfs[idx].text)
        result3.append(tmns[idx].text)
        result4.append(tmxs[idx].text)
        result5.append(rnsts[idx].text)

print(result1[:5])
print(result2[:5])
result=zip(result1,result2,result3,result4,result5) # 리스트의 같은 인덱스끼리 저장
rows=list(result)
rows
```

```python
import pandas as pd
df=pd.DataFrame(rows,columns=['일시','날씨','최저기온','최고기온','강수확률'])
df.head()
```

```python
df.to_csv('d:/data/weather/data.csv', encoding='ms949', index=False) # csv 파일로 저장
```

```python
from bs4 import BeautifulSoup
html="""
<html><body>
<div id="main">
    <h1>도서 목록</h1>
    <ul class="items">
        <li>빅데이터</li>
        <li>HTML</li>
        <li>Python</li>
    </ul>
</div>
</body></html>
"""

bs=BeautifulSoup(html, 'html.parser')
h1=bs.select_one("div#main > h1").text # id가 main인 div 태그의 자손 h1 태그
print("h1 :",h1)

li_list=bs.select("div#main > ul.items > li") # id가 amin인 div 태그 → class가 items인 ul 태그 → li 태그들
for li in li_list:
    print("li :",li.text)
```

```python
# 환율 정보 수집
from bs4 import BeautifulSoup
import urllib.request as req
url="http://finance.naver.com/"
res=req.urlopen(url)
bs=BeautifulSoup(res, 'html.parser')

result1=[]
result2=[]
moneys=bs.select('#content > div.article2 > div.section1 > div.group1 > table > tbody > tr > th > a')
for money in moneys:
    result1.append(money.text.strip())
exchange=bs.select('#content > div.article2 > div.section1 > div.group1 > table > tbody > tr > td:nth-child(2)')
for ex in exchange:
    result2.append(ex.text)

result=list(zip(result1, result2))
result
```

```python
# 옥션 상품 분류 코드 가져오기
from bs4 import BeautifulSoup
import urllib.request as req
import urllib.request
from urllib.request import urlopen

codes=[]
for num in range(51,53):
    url=f"http://www.auction.co.kr/category/category{num}.html"
    print(url)
    res=req.urlopen(url)
    bs=BeautifulSoup(res, 'html.parser', from_encoding='ms949')
    category=bs.select('#navCat > div > ul > li > a')
    for cate in category:
        name=cate.text
        href=cate.attrs['href'].split('=')[1]
        codes.append(href)
        print('상품 분류명 :',name)
        print('분류 코드 :',href+'\n')
```

```python
for cate in codes:
    html=urlopen('http://browse.auction.co.kr/list?category='+cate)
    bs=BeautifulSoup(html.read(), 'html.parser')
    tags=bs.find_all('img')
    for img in tags:
        if 'http:' not in img['src']:
            src='http:'+img.attrs['src']
        else:
            src=img.attrs['src']
        print(src)
        name=src.rsplit('/',1)[-1]
        urllib.request.urlretrieve(src, "d:/data/images/"+name)
```
