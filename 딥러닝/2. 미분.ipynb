```python
# x를 중심으로 전후의 차분을 계산 (중심차분 또는 중앙차분)
def numerical_diff(f,x):
    h=1e-4 # 0.0001
    return (f(x+h)-f(x-h))/(2*h)
```

```python
# 아래와 같은 이차함수가 있을 때
def function_1(x):
    return 0.01*x**2+0.1*x
```

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.arange(0,20,0.1)
y=function_1(X)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(X,y)
plt.show()
```

```python
print(numerical_diff(function_1,1)) # x=1 일 때 미분계수
print(numerical_diff(function_1,5)) # x=5 일 때 미분계수
print(numerical_diff(function_1,10)) # x=10 일 때 미분계수
```

```python
def tangent_line(f,x):
    d=numerical_diff(f,x)
    y=f(x)-d*x
    return lambda t: d*t+y

X=np.arange(0,20,0.1)
y=function_1(X)
plt.xlabel('x')
plt.ylabel('f(x)')

# x가 5일때 y=0.2
# x=5에서의 접선
tf=tangent_line(function_1, 5)
y2=tf(X)

plt.plot(X, y) # 이차함수
plt.plot(X, y2) # 접선
plt.scatter(5,0.75, c='r') # 접점
plt.show()
```

```python
# x=10에서의 접선
tf=tangent_line(function_1, 10)
y2=tf(X)

plt.plot(X, y) # 이차함수
plt.plot(X, y2) # 접선
plt.scatter(10,2, c='r') # 접점
plt.show()

# 경사하강법 : 기울기를 조금씩 변화시켜 가면서 가장 손실률이 낮은 값을 찾는 방법
#             현 위치에서 기울어진 방향으로 일정 거리만큼 이동 - 기울기를 구한다
#             오차가 최소인 기울기를 찾는 방법. 갱신하는 양을 학습률이라고 한다.
#             한번의 학습에 얼만큼 학습을 할지 정하는 것, 매개변수 값을 얼마나 갱신할 것인가?
```
