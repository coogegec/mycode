```python
# 전처리된 파일
data2=pd.read_csv('d:/data/insurance/insurance2.csv')
data.head()
```

```python
data2.corr()['charges'].sort_values()
```

```python
# 명목형 변수는 제외하고 연속형 변수만 독립변수로 선택
import statsmodels.api as sm

df2=data2.iloc[:,[0,2,3]]
print(df2.head())
y2=data2.iloc[:,[9]] # 종속변수
print(y2.head())

model2=sm.OLS(y2, df2)
result2=model2.fit()

print('회귀계수\n',result2.params)
print('R-squared\n',result2.rsquared) # 회귀모델의 설명력 : 59.5%
print('P values\n',result2.pvalues)
print(result2.summary()) # 요약 결과
```

```python
# 유의하지 않은 children 변수 제거
df3=data2.iloc[:,[0,2]]
print(df3.head())
y3=data2.iloc[:,[9]] # 종속변수
print(y3.head())

model3=sm.OLS(y3, df3)
result3=model3.fit()

print('회귀계수\n',result3.params)
print('R-squared\n',result3.rsquared) # 회귀모델의 설명력 : 95.8%
print('P values\n',result3.pvalues)
print(result3.summary()) # 요약 결과
```
